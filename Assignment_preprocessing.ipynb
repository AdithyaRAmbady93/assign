## 5.Check Outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

A boxplot is a standardized way of displaying the dataset based on a five-number summary: the minimum, the maximum, the sample median, and the first and third quartiles.

It also helps us to detect outliers using IQR (Inter Quantile Range) method.
# 1.Check and Handle outliers in column Age


```python
plt.figure(figsize = (10,6))
sns.boxplot(tit.Age)
plt.title('Age',size=18)
plt.show()
```

```python
Q1 = tit['Age'].quantile(0.25)
Q3 = tit['Age'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
tit[(tit['Age']< Q1-1.5* IQR) | (tit['Age']> Q3+1.5* IQR)]
```

From above, We can easily detect outliers with a boxplot or using python code

# 2.Check and Handle outliers in column Fare

```python
plt.figure(figsize = (10,6))
sns.boxplot(tit.Fare)
plt.title('Fare',size=18)
plt.show()
```

from the boxplot above, the black circular points which are indicated by show the presence of extreme values in the variable.

```python
Q1 = tit['Fare'].quantile(0.25)
Q3 = tit['Fare'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

The code below print the outliers and sets the 25th and 75th percentile of the ‘Fare’ variable respectively .

```python
tit[(tit['Fare']< Q1-1.5* IQR) | (tit['Fare']> Q3+1.5* IQR)]
```
